```python
#This codes requires you to store the data at /data/XXXX.csv
oil = pd.read_csv('data/oil.csv')

oil['date'] = pd.to_datetime(oil['date'])
StartDate = date(2013,1,1)
oil['dif'] = oil['date'] - StartDate

oil['Days'] = 0
length = len(oil)
for i in range (0,length):
    oil['Days'].loc[i] = float(oil['dif'].loc[i].days)
oil.drop('dif', axis=1, inplace= True)

oil_new = oil[np.isfinite(oil['dcoilwtico'])]

plt.plot(oil_new.Days, oil_new.dcoilwtico)
plt.xlabel('Days from 2013 Jan-01')
plt.ylabel('Oil Price')
plt.show()
```
